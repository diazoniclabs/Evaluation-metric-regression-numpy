**SAMPLE OF OUTPUT VALUES**

```python
import numpy as np

y = np.array([100,200,150,250,300])
```

```python
y_pred = np.array([110,210,140,245,290])
```

**MEAN SQUARED ERROR**

```python
# From Numpy
mse = np.mean(np.square(np.subtract(y,y_pred)))
mse
```

```python
# From Sklearn
from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_pred)
```

**MEAN ABSOLUTE ERROR**

```python
# From Numpy
mse = np.mean(np.abs(np.subtract(y,y_pred)))
mse
```

```python
# From Sklearn
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,y_pred)
```

**R2 SCORE**

image.png

```python
# From Numpy
sse = np.sum(np.square(np.subtract(y,y_pred)))
sse
```

```python
sst = np.sum(np.square(np.subtract(y,np.mean(y))))
sst
```

```python
r2 = 1-(sse/sst)
r2
```

```python
# From Sklearn
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```
